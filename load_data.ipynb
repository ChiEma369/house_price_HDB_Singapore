```python
import pandas as pd
import numpy as np
import re

DATA_PATH = '/content/drive/MyDrive/ColabNotebooks/HDBprices/data-processed/hdb_resale_2021-2025.csv'

df = pd.read_csv(DATA_PATH)

print(df.shape)
df.head()
```

```python
df.info()
```

# Chuẩn hóa dữ liệu file csv mới
- vì model không chạy được với dữ liệu chữ, nên lấy trung bình tầng (strorey_mid)
- đổi dữ liệu về dạng tháng thập phân



```python
def get_storey_mid(storey_range):
    numbers = re.findall(r'\d+', str(storey_range))

    if len(numbers) == 2:
        low = int(numbers[0])
        high = int(numbers[1])
        return (low + high) / 2   #trung bình

    return np.nan  #nếu dữ liệu lạ trả về giá trị thiếu

df['storey_mid'] = df['storey_range'].apply(get_storey_mid)

df[['storey_range', 'storey_mid']].head(10)
```

```python
def lease_to_years(text):
  text = str(text)

  year = re.search(r'(\d+)\s*year', text)
  month = re.search(r'(\d+)\s*month', text)

  years = int(year.group(1)) if year else 0
  months = int(month.group(1)) if month else 0

  return years + months / 12

df['lease_years'] = df['remaining_lease'].apply(lease_to_years)

df[['remaining_lease', 'lease_years']].head(10)
```

```python
df[['storey_range', 'storey_mid', 'remaining_lease', 'lease_years']].sample(10, random_state=42)
```

- cần dự đoán resale_price
- gộp danh sách số và chữ vào features


```python
target = 'resale_price'

numeric_features = ['floor_area_sqm', 'lease_age', 'lease_years', 'storey_mid', 'year']
categorical_features = ['town', 'flat_type', 'flat_model']

features = numeric_features + categorical_features

x = df[features]
y = df[target]

print("X shape:", x.shape)
print("Y shape:", y.shape)

x.head()
```

```python
#kiểm tra null
x.isnull().sum()
```

# Chia train/test

```python
from sklearn.model_selection import train_test_split

#20% test, 80% train, cố định chia dữ liệu 42
x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.2, random_state=42
)
print("Train size:", x_train.shape)
print("Test size:", x_test.shape)
```

# Chuẩn hóa cột số và mã hóa cột chữ
- StandardScaler_ số
- OneHotEnncoder_ chữ
- ColumTranformer _ gom việc

```python
from sklearn.compose import ColumnTransformer
# nhóm công cụ preprocessing biến đổi raw feture thành dạng phù hợp model
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
```

- các dữ liệu về chung đơn vị, ép nhỏ các trọng số w quá lớn_hàm phạt Ridge giải quyết
- 'SimpleImputer': phương trình hồi quy cần ma trận hoàn hảo -> fill các dòng dữ liệu bị thiếu = số trung vị/giá trị xuất hiện nhiều nhất

```python
numeric_trainsformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_trainsformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
])
```

# Train linear regression

```python
from sklearn.linear_model import LinearRegression

linear_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])
#học từ dữ liệu train
linear_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = linear_model.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Linear Regression results")
print("MAE", mae)
print("MSE", mse)
print("R2", r2)    #dễ hiểu hơn vì cùng giá trị với giá nhà, càng gần 1 càng tốt
```

# Train Ridge Regression
(so sánh thêm)

```python
from sklearn.linear_model import Ridge

ridge_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', Ridge(alpha=1.0))      #phat muc alpha, model cang lon thi a cang nho
])
ridge_model.fit(x_train, y_train)
ridge_pred = ridge_model.predict(x_test)

rid_mae = mean_absolute_error(y_test, ridge_pred)
rid_mse = mean_squared_error(y_test, ridge_pred)
rid_rmse = np.sqrt(mse)
rid_r2 = r2_score(y_test, ridge_pred)

print("Ridge Regression results")
print("MAE", mae)
print("MSE", mse)
print("R2", r2)
```

```python
results = pd.DataFrame({
    'model': ['Linear Regression', 'Ridge Regression'],
    'MAE': [mae, rid_mae],
    'RMSE': [rmse, rid_rmse],
    'R2': [r2, rid_r2]
})

results
```

Định nghĩa hàm Evaluate

```python
def evaluate_model(model, x_test, y_test, model_name):
    y_pred = model.predict(x_test)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    print(f"--- {model_name} results ---")
    print("MAE:", mae)
    print("RMSE:", rmse)
    print("R2:", r2)

    return {'model': model_name, 'MAE': mae, 'RMSE': rmse, 'R2': r2}
```

Nhận xét: Linear Regression đạt kết quả tương đương Ridge trong baseline đầu tiên, cho thấy mô hình tuyến tính chưa bị ảnh hưởng nhiều bởi hệ số quá lớn ở bộ feature hiện tại.

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', RandomForestRegressor(
        n_estimators=100,
        max_depth=20,
        random_state=42,
        n_jobs=-1
    ))
])

rf_model.fit(x_train, y_train)

rf_result = evaluate_model(
    rf_model,
    x_test,
    y_test,
    'Random Forest'
)

rf_result
```

Lưu model Random Forest

```python
import joblib

model_path = PROCESSED_DIR + '/rf_model_v1.pkl'

# 1. Lệnh LƯU model 
joblib.dump(rf_model, model_path)
print(f"Đã lưu model tại: {model_path}")
```

```python
rf_model = joblib.load(model_path)
print("Đã tải model thành công!")
```

So sánh hiệu suất các model

```python
# 1. lưu các thông số của 2 model trc
linear_res = evaluate_model(linear_model, x_test, y_test, 'Linear Regression')
ridge_res = evaluate_model(ridge_model, x_test, y_test, 'Ridge Regression')

# 2. Tạo bảng so sánh tổng hợp
comparison_df = pd.DataFrame([linear_res, ridge_res, rf_result])

print("\nBẢNG SO SÁNH HIỆU SUẤT 3 MÔ HÌNH:")
display(comparison_df.sort_values(by='R2', ascending=False))
```
